```python

import pandas as pd
import numpy as np
```

```python
df_train=pd.read_csv('/content/train.csv')
df_train.head()
```

```python
df_test=pd.read_csv('/content/test.csv')
df_test.head()
```

```python
df = pd.concat([df_train, df_test], ignore_index=True)
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Gender']=df['Gender'].apply(lambda x: 0 if x == 'F' else 1)
```

```python
df.head()
```

```python
## Handle categorical feature Age
df['Age'].unique()
```

```python
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
df.head()
```

```python
df['City_Category'].unique()
```

```python
df['City_Category']=df['City_Category'].map({'A':1,'B':2,'C':3})
```

```python
df.isnull().sum()
```

```python
##replacing nan values in purchase categpry 2 and purchase category 3
df['Product_Category_1'].unique()
```

```python
##Best way to replace a missing feature is with mode for categorical values
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].astype(str).astype(int)
```

```python
df.info()
```

```python

import seaborn as sns
sns.pairplot(df)
```

```python
df_train=df[~df['Purchase'].isnull()]
df_test=df[df['Purchase'].isnull()]
```

```python
X=df_train[df_train.columns[1:10]]
X.head()
```

```python
y=df_train['Purchase']
y.head()
```

```python
#feature scaling

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn import preprocessing

```

```python

# creating a regression model
model = LinearRegression()
```

```python

# fitting the model
model.fit(X_train,y_train)
```

```python
# making predictions
predictions = model.predict(X_test)
```
