```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
df
```

```python
df.columns
```

```python
country_names= df.Country.value_counts().index
country_names
```

```python
country_val=df.Country.value_counts().values
```

```python
plt.pie(country_val[:4],labels=country_names[:4],autopct="%1.2f%%")
```

```python
df.columns
```

```python
ratings=df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['white', 'red', 'orange', 'yellow', 'green', 'green'])
```

```python
sns.countplot(x='Rating color',data=ratings)
```

```python
dfs=df[df['Rating color']== 'White'].groupby('Country').size().reset_index()
dfs
```
